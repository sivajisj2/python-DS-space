```python
import re

raw_text = """
301 SOS0392332
Name : VENKATNRSRAJU UPPALAPATI
Fathers Name: RAMARAJU
House Number : 4/91
Age : 73 Gender: Male  Photo
Available   Name : GURUPRASAD UPPALAPATI Fathers
Name: VENKATA NARASA RAJU House
Number: 4/91
Age : 34 Gender: Male  Photo


305
SOS1237098
306 SOS1317619
307 SOS1317627
Name : VAISHALINI NARAPARAJU
Husbands Name: GURUPAVAN RAJU
UPPALAPATI
House Number : 4/92
Age : 27 Gender: Female  Photo
Available   Name : LAKSHMI DEVAM MA UPPALAPATI
Husbands Name: CHINNAGURAVARAJU House
Number: 4/92
Age : 63 Gender : Female  Photo
Available   Name : CHINNAGURAVARAJU UPPALAPATI
Fathers Name: RAMRAJU
House Number: 4/92
Age : 68 Gender: Male  Photo
Available
           Assembly Constituency No and Name : 127 -KODUR (SC)
Section No and Name 2 -PATHIGUNTAVARIPALLI  Part No.: 4
Age as on 01 -01-2024  #/#1 Modified as per supplement Date of Publication : - 22-01-2024  Total Pages 16 - Page 15
 308
SOS0802620  309
SOS1317635
KXD0 749622
Name : SAVITHRAMMA UPPALAPATTI
Husbands Name: Guravaraju Uppalapati House
Number: 4/92A
Age : 73 Gender: Female  Photo
Available   Name : RAMAKKA ATMAKURU Husbands Name:
BALARAMARAJU House Number: 4/93
Age : 85 Gender : Female  Photo
Available   Name : EESHVARAMMA ATMAKURU
Husbands Name: 1IRUPALARAJU
House Number: 4/93
Age : 48 Gender: Female  Photo
Available


| 311 SOS0443754  312
KXD0750414  1
SOS1912666
Name : THIRUPALU RAJU ATHMAKURU
Fathers Name: BALARAMARAJU ATHMAKURU
House Number: 4/93
Age : 51 Gender: Male  Photo
Available   Name : NARAAYANA RAJU KOTCHERLA
Fathers Name: CHINNAOBULARAJU House
Number: 4/95
Age : 89 Gender : Male  Photo
Available   Name : Baghya Lakshmi Kocherla Fathers Name:
Anjaneya Raju Kocherla House Number: 4/95
Age : 20 Gender: Female  Photo
Availab le
"""

# Function to extract relevant information from each block of text
def extract_info(block):
    serial_number_match = re.search(r'\bSOS\d+\b', block)
    serial_number = serial_number_match.group() if serial_number_match else None
    name_match = re.search(r'Name : (.+?)(?:\s+Fathers Name|\s+Husbands Name)', block)
    name = name_match.group(1).strip() if name_match else None
    father_or_husband_name_match = re.search(r'(?:Fathers|Husbands) Name: (.+?)(?:\s+Name|$)', block)
    father_or_husband_name = father_or_husband_name_match.group(1).strip() if father_or_husband_name_match else None
    house_number_match = re.search(r'House Number : ([\w\-/]+)(?:\s+Age|$)', block)
    house_number = house_number_match.group(1).strip() if house_number_match else None
    age_match = re.search(r'Age : (\d+)', block)
    age = age_match.group(1).strip() if age_match else None
    gender_match = re.search(r'Gender : (\w+)', block)
    gender = gender_match.group(1).strip() if gender_match else None
    return {
        "Serial Number": serial_number,
        "Name": name,
        "Father/Husband Name": father_or_husband_name,
        "House Number": house_number,
        "Age": age,
        "Gender": gender
    }



# Split the text into blocks
blocks = re.split(r'\b\d+\b', raw_text.strip())

# Process each block to extract information
data = [extract_info(block) for block in blocks if block.strip()]

# Print the extracted data
for entry in data:
    print(entry)
```
